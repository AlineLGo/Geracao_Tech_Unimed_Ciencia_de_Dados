```python
# Importando a biblioteca
import pandas as pd
```

```python
# Leitura dos arquivos
df1 = pd.read_excel("/content/drive/MyDrive/Mini Curso- Analise de Dados/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Mini Curso- Analise de Dados/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Mini Curso- Analise de Dados/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Mini Curso- Analise de Dados/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Mini Curso- Analise de Dados/Salvador.xlsx")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df1.head()
```

```python
df2.head()
```

```python
# Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
# Exibindo as 5 primeiras linhas
df.head()
```

```python
# Exibindo as 5 últimas linhas
df.tail()
```

```python
df.sample(5)
```

```python
df["LojaID"] = df["LojaID"].astype("object") # Alterando o tipo de dado da coluna Região
```

```python
df.dtypes
```

```python
df.isnull().sum() # Consultando linhas com valores faltantes
```

```python
#Substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(),inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True) # inplace substitui o objeto em memória
```

```python
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

Criar novas colunas

```python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest
df.nlargest(3, "Receita") # retorna as 3 linhas com maiores receitas
```

```python
df.nsmallest(3, "Receita")# retorna as 3 linhas com menores receitas
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10) #ascending =False - retorna do maior para o menor
```

**Trabalhando com datas**

```python
#Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype('int64')
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias
df['diferença_dias'] = df['Data'] - df['Data'].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df['trimestre_venda'] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19
```
