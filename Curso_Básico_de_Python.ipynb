**Passo a passo de solução do Desafio**

```python
# Lógica de Programaação

# Passo 0 - Entender o desafio que se quer resolver

# Passo 1 - Percorrer todos os arquivos da pasta base de daddos (Pasta Vendas)
import os
import pandas as pd

lista_arquivo = os.listdir("/content/drive/MyDrive/Mini Curso- Analise de Dados/Curso Básico de Python/Curso Básico de Python (1)/Vendas")
display(lista_arquivo)

```

```python
tabela_total = pd.DataFrame()

# Passo 2 - Importar as bases de dados de vendas
for arquivo in lista_arquivo:
  if "Vendas" in arquivo:

  # f = significa que aceita formatação dinâmica.
    tabela = pd.read_csv(f"/content/drive/MyDrive/Mini Curso- Analise de Dados/Curso Básico de Python/Curso Básico de Python (1)/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela)  # append - adiciona informações dentro da tabela



# Passo 3 - Tratar / Compilar as bases de dados
display(tabela_total)

# Passo 4 - Calcular o produto mais vendido (em quantidade)

# Passo 5 - Calcular o produto que mais faturou (em faturamento)

# Passo 6 - Calcular a loja/cidade que mais vendeu (em faturamento) - cri
```

```python
# Passo 4 - Calcular o produto mais vendido (em quantidade)
tabela_produtos = tabela_total.groupby("Produto").sum() # aqui agrupa a coluna de produto, para aparecer os podutos uma única vez, fazendo a soma
tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by="Quantidade Vendida", ascending=False) # aqui diz qual as colunas que vão ser visualizada,ordenando de mais vendido para o menos vendido
display(tabela_produtos)
```

```python
# Passo 5 - Calcular o produto que mais faturou (em faturamento)
# criar uma nova coluna de faturamento para adicionar na tabela
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario'] # aqui faz o calculo do faturamento

tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by="Faturamento", ascending=False)
display(tabela_faturamento)
```

```python
# Passo 6 - Calcular a loja/cidade que mais vendeu (em faturamento) - criação de gráfico/dasboard
tabela_lojas = tabela_total.groupby("Loja").sum()
tabela_lojas = tabela_lojas[['Faturamento']]
display(tabela_lojas)
```

```python
# Criar gráfico/ dashboard
import plotly.express as px

grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y='Faturamento')
grafico.show()
```
